### This tutorial shows lots of different pulses and their Fourier transforms

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
from scipy.fftpack import fft as fft

Fs = 200                         # sampling rate
Ts = 1.0/Fs                      # sampling interval
t = np.arange(0,1,Ts)            # time vector
ff = 5                           # frequency of the signal

nPulse = 20
y = np.ones(nPulse)
y = np.append(y, np.zeros(Fs-nPulse))
print(y)

fig, ax = plt.subplots(figsize=(6,6), dpi=144)
plt.subplot(2,1,1)
plt.plot(t,y,'k-')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,1,2)
n = len(y)                       # length of the signal
N = int(n)
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
freq = frq[1:int(N/2)]#[range(int(n)/2)]           # one side frequency range

Y = fft(y)/int(n)              # fft computing and normalization
Y = Y[1:int(N/2)]#[range(int(n)/2)]

plt.plot(freq, abs(Y), 'r-')
plt.xlabel('freq (Hz)')
plt.ylabel('|Y(freq)|')

plt.show()
```

## Pulse width modulated sine wave

Here the pulse width is proportional to the phase of the sine wave

When the sine wave is high, the pulses are long<br>
When the sine wave is low, the pulses are short

```python
# pulse width modulated sine wave


sig = np.sin(2 * np.pi * t)
pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
y = pwm

fig, ax = plt.subplots(figsize=(6,6), dpi=144)  
plt.subplot(3, 1, 1)
plt.plot(t, sig)
plt.subplot(3, 1, 2)
plt.plot(t, pwm)
plt.ylim(-1.5, 1.5)


plt.subplot(3,1,3)
n = len(y)                       # length of the signal
N = int(n)
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
freq = frq[1:int(N/2)]#[range(int(n)/2)]           # one side frequency range

Y = fft(y)/int(n)              # fft computing and normalization
Y = Y[1:int(N/2)]#[range(int(n)/2)]           

plt.plot(freq, abs(Y), 'r-') 
```

```python
from scipy.signal import sweep_poly
p = np.poly1d([0.025, -0.36, 1.25, 2.0])
t = np.linspace(0, 10, 5001)
w = sweep_poly(t, p)
y = w

fig, ax = plt.subplots(figsize=(6,6), dpi=144)    
plt.subplot(2,1,1)
plt.plot(t,y,'k-')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,1,2)
n = len(y)                       # length of the signal
N = int(n)
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
freq = frq[1:int(N/2)]#[range(int(n)/2)]           # one side frequency range

Y = fft(y)/int(n)              # fft computing and normalization
Y = Y[1:int(N/2)]#[range(int(n)/2)]           
           
plt.plot(freq, abs(Y), 'r-')
plt.xlim(0,10)
```

```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 1500)
w = chirp(t, f0=6, f1=1, t1=10, method='linear')
plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()
```

```python
fs = 7200
T = 4
t = np.arange(0, int(T*fs)) / fs
```

```python
def plot_spectrogram(title, w, fs):
    ff, tt, Sxx = spectrogram(w, fs=fs, nperseg=256, nfft=576)
    fig, ax = plt.subplots()
    ax.pcolormesh(tt, ff[:145], Sxx[:145], cmap='gray_r',
                  shading='gouraud')
    ax.set_title(title)
    ax.set_xlabel('t (sec)')
    ax.set_ylabel('Frequency (Hz)')
    ax.grid(True)
```

```python
w = chirp(t, f0=1500, f1=250, t1=T, method='quadratic')
plot_spectrogram(f'Quadratic Chirp, f(0)=1500, f({T})=250', w, fs)
plt.show()
```

```python
w = chirp(t, f0=1500, f1=250, t1=T, method='quadratic',
          vertex_zero=False)
plot_spectrogram(f'Quadratic Chirp, f(0)=1500, f({T})=250\n' +
                 '(vertex_zero=False)', w, fs)
plt.show()
```

```python
w = chirp(t, f0=1500, f1=250, t1=T, method='hyperbolic')
plot_spectrogram(f'Hyperbolic Chirp, f(0)=1500, f({T})=250', w, fs)
plt.show()
```
